```python
import numpy as np
from cmdstanpy import CmdStanModel
```

Excercise 1

```python
F = 7
L = 9
N = (L+F)*100

model = CmdStanModel(stan_file='code_1.stan')
data = {'N': N}
result = model.sample(data=data, chains = 4)
```

```python
result.draws_pd()
```

```python
model_2 = CmdStanModel(stan_file='code_2.stan')
data_2 = {'N': N}
result_2 = model_2.sample(data=data_2, chains = 4)
```

```python
result_2.draws_pd()
```

Excercise 2

```python
y = result_2.stan_variable("y")[0].tolist()
```

```python
model_3 = CmdStanModel(stan_file='code_3.stan')

data = {'N': N, 'y': y}
result_3 = model_3.sample(data=data, chains = 4)
```

```python
import arviz as az
arviz_result_3 = az.from_cmdstanpy(posterior=result_3)
```

```python
arviz_result_3
```

```python
model_4 = CmdStanModel(stan_file='code_4.stan')
result_4 = model_4.sample(data=data, chains = 4)
arviz_result_4 = az.from_cmdstanpy(posterior=result_4)
arviz_result_4
```

```python
compare_dict = {"code_3": arviz_result_3, "code_4": arviz_result_4}
az_compare = az.compare(compare_dict)
az_compare
```

```python
az.plot_compare(az_compare)
```

```python
compare_dict = {"code_3": arviz_result_3, "code_4": arviz_result_4}
az_compare = az.compare(compare_dict, ic="waic")
az_compare
```

```python
az.plot_compare(az_compare)
```

Excercise 3

```python
# K = 1

model_5 = CmdStanModel(stan_file='code_5.stan')

X = result_2.stan_variable("X")[0][:, 0:1]
K = 1
data = {"N": N, "y":y, "X": X, "K": K}

result_5 = model_5.sample(data=data, chains=4)
arviz_result_5 = az.from_cmdstanpy(posterior=result_5)
arviz_result_5
```

```python
# K = 2

K = 2
X = result_2.stan_variable("X")[0][:, 0:K]
data = {"N": N, "y":y, "X": X, "K": K}

result_5_K_2 = model_5.sample(data=data, chains=4)
arviz_result_5_K_2 = az.from_cmdstanpy(posterior=result_5_K_2)
arviz_result_5_K_2
```

```python
# K = 3

K = 3
X = result_2.stan_variable("X")[0][:, 0:K]
data = {"N": N, "y":y, "X": X, "K": K}

result_5_K_3 = model_5.sample(data=data, chains=4)
arviz_result_5_K_3 = az.from_cmdstanpy(posterior=result_5_K_3)
arviz_result_5_K_3
```

Compare models

```python
compare_dict = {"K = 1": arviz_result_5, "K = 2": arviz_result_5_K_2, "K = 3": arviz_result_5_K_3}
az_compare = az.compare(compare_dict, ic="waic")
az_compare
```

```python
az.plot_compare(az_compare)
```

```python
az_compare = az.compare(compare_dict)
az_compare
```

```python
az.plot_compare(az_compare)
```
